```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataframe = pd.read_csv("/kaggle/input/houses1/houses - houses.csv")
dataframe.head(n=10)
```

```python
dataframe["Location"] = dataframe["Location"].apply(lambda x:1 if x == "Ontario" else 0)

dataframe.head() 
```

lambda functions are more of a one line loops, anonymus function, one use 

```python
plt.scatter(dataframe["Square Feet"], dataframe["House Prices"], c=dataframe["Location"])
plt.xlabel("Square Feet")
plt.ylabel("House Prices")

cbar = plt.colorbar()
```

plt.scatter(dataframe["Square Feet"] - shows the type of plot and the value of x axis as square feet.
dataframe["House Prices"], - shows y asix as house prices 
 c=dataframe["Location"]) -  c is colour and it represents the 2 places ontario and texas.

```python
features = dataframe[["Square Feet", "House Prices"]]
labels = dataframe[["Location"]]
```

```python
features_train, features_test, labels_train, labels_test = train_test_split (features, labels, train_size = 0.8, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(features_train, labels_train)
```

```python
predictions = model.predict(features_test)
```

```python
plt.scatter(labels_test, predictions)
plt.xlabel("Actual Location")
plt.ylabel("predicted Location")
```

```python
print(model.score(features_test, labels_test))
```
