```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataframe = pd.read_csv("/kaggle/input/iris-data/iris_data (1).csv")
dataframe.head(n=10)
```

```python
dataframe["Species"] = dataframe["Species"].apply(lambda x:1 if x == "Iris-setosa" else 0)

dataframe.head() 
```

```python
plt.scatter(dataframe["PetalLengthCm"], dataframe["PetalWidthCm"], c=dataframe["Species"])
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")

cbar = plt.colorbar()
```

```python
features = dataframe[["PetalLengthCm", "PetalWidthCm"]]
labels = dataframe[["Species"]]
```

```python
features_train, features_test, labels_train, labels_test = train_test_split (features, labels, train_size = 0.8, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(features_train, labels_train)
```

```python
predictions = model.predict(features_test)
```

```python
plt.scatter(labels_test, predictions)

plt.xlabel("Actual Species")
plt.ylabel("predicted Species")
```

```python
print(model.score(features_test, labels_test))
```
